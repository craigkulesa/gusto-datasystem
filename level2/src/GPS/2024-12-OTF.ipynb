```python
from tqdm import tqdm
import matplotlib.pyplot as plt

def make_otf_map(glon_min,glon_max,arcmin_spacing,direction,line,mixer,ax=None):
    if line=="CII":
        linefreq=1900537
    else:
        line="NII"
        linefreq=1461132
    if ax is None:
        ax = plt.gca()
    GLON_MIN, GLON_MAX=(glon_min, glon_max)
    if GLON_MAX < 50:
        GLON_MAX = GLON_MAX + 360
    GLAT_MIN, GLAT_MAX=(0.5, 1.0)
    SPACING=arcmin_spacing/60
    nx, ny = (int((GLON_MAX-GLON_MIN)/SPACING)+1 , int((GLAT_MAX-GLAT_MIN)/SPACING)+1)
    x = np.linspace(GLON_MIN,GLON_MAX,nx)
    y = np.linspace(GLAT_MIN,GLAT_MAX,ny)
    xv , yv = np.meshgrid(x,y)
    
    otf_grid=np.zeros_like(xv)

    otfsel=(otf['TARMS']<20) & (otf['PWR']>-10) & (otf['PWR']<1000) & (otf['LINEFREQ']==linefreq) & (otf['SCANDIR']==direction) & (otf['MIXER']==mixer)
    otf2=otf[otfsel]

    for i, x in tqdm(enumerate(xv)):
        for j, y in enumerate(x):
            otfsel=(otf2['GLON']>xv[i,j]) & (otf2['GLON']<(xv[i,j]+SPACING)) & (otf2['GLAT']>yv[i,j]) & (otf2['GLAT']<(yv[i,j]+SPACING))
            otf3=otf2[otfsel]
            otf_grid[i,j]=np.ma.mean(otf3['PWR'])
    hist = ax.pcolormesh(xv,yv,otf_grid,vmax=100)
    ax.title.set_text("%s mix:%d dir:%s" % (line, mixer, direction))
    ax.set_aspect('equal', adjustable='datalim')
    plt.colorbar(hist,ax=ax)
    return hist
```

```python
from tqdm import tqdm
import matplotlib.pyplot as plt

def make_otf_map_tarms(glon_min,glon_max,arcmin_spacing,direction,line,mixer,ax=None):
    if line=="CII":
        linefreq=1900537
    else:
        line="NII"
        linefreq=1461132
    if ax is None:
        ax = plt.gca()
    GLON_MIN, GLON_MAX=(glon_min, glon_max)
    if GLON_MAX < 50:
        GLON_MAX = GLON_MAX + 360
    GLAT_MIN, GLAT_MAX=(-2.0, 2.0)
    SPACING=arcmin_spacing/60
    nx, ny = (int((GLON_MAX-GLON_MIN)/SPACING)+1 , int((GLAT_MAX-GLAT_MIN)/SPACING)+1)
    x = np.linspace(GLON_MIN,GLON_MAX,nx)
    y = np.linspace(GLAT_MIN,GLAT_MAX,ny)
    xv , yv = np.meshgrid(x,y)
    
    otf_grid=np.zeros_like(xv)

    if direction=="both":
        otfsel=(otf['LINEFREQ']==linefreq) & (otf['MIXER']==mixer)
    else:
        otfsel=(otf['LINEFREQ']==linefreq) & (otf['SCANDIR']==direction) & (otf['MIXER']==mixer)
    otf2=otf[otfsel]

    for i, x in tqdm(enumerate(xv)):
        for j, y in enumerate(x):
            otfsel=(otf2['GLON']>xv[i,j]) & (otf2['GLON']<(xv[i,j]+SPACING)) & (otf2['GLAT']>yv[i,j]) & (otf2['GLAT']<(yv[i,j]+SPACING))
            otf3=otf2[otfsel]
            otf_grid[i,j]=np.ma.mean(otf3['TARMS'])
    hist = ax.pcolormesh(xv,yv,otf_grid,vmax=100)
    ax.title.set_text("%s mix:%d dir:%s" % (line, mixer, direction))
    ax.set_aspect('equal', adjustable='datalim')
    ax.set_xlim((GLON_MAX,GLON_MIN))
    plt.colorbar(hist,ax=ax)
    return hist
```
